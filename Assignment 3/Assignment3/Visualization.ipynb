```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from fastmri.data.subsample import create_mask_for_mask_type
```

```python
h5f = h5py.File("C:/Users/jonas/Downloads/file061.h5", 'r')
h5f_test = h5py.File("C:/Users/jonas/Downloads/file125.h5", 'r')
```

```python
print(h5f.keys())
```

```python
h5f['kspace'].shape
```

```python
plt.figure(figsize=(12, 6))

# Plot the real part
plt.subplot(1, 2, 1)
plt.title('Real Part')
plt.imshow(h5f['kspace'][18, 0, :, :].real, cmap='gray')
plt.colorbar()

# Plot the imaginary part
plt.subplot(1, 2, 2)
plt.title('Imaginary Part')
plt.imshow(abs(np.fft.fftshift(np.fft.ifft2(h5f['kspace'][18, 0, :, :]))), cmap='gray')
plt.colorbar()

plt.show()
```

```python
slice = 18

plt.figure(figsize=(24, 6))

# Plot the real part
plt.subplot(1, 4, 1)
plt.title('Real Part')
plt.imshow(np.log(np.fft.fftshift(np.fft.fft2(h5f['kspace'][slice, 0, :, :]))).real, cmap='gray')
plt.colorbar()

# Plot the imaginary part
plt.subplot(1, 4, 2)
plt.title('Imaginary Part')
plt.imshow(np.log(np.fft.fftshift(np.fft.fft2(h5f['kspace'][slice, 0, :, :]))).imag, cmap='gray')
plt.colorbar()

# Plot the magnitude
plt.subplot(1, 4, 3)
plt.title('Magnitude')
plt.imshow(np.abs(np.fft.fftshift(np.fft.fft2(h5f['kspace'][slice, 0, :, :]))), cmap='gray')
plt.colorbar()

# Plot the phase
plt.subplot(1, 4, 4)
plt.title('Magnitude')
plt.imshow(np.angle(np.fft.fftshift(np.fft.fft2(h5f['kspace'][slice, 0, :, :]))), cmap='gray')
plt.colorbar()


plt.show()
```

```python
mask = create_mask_for_mask_type(
    'random', 
    center_fractions=[0.08], 
    accelerations=[4]
)

kspace_mask = mask(h5f['kspace'][slice].shape)

# Apply the mask to the k-space data
masked_kspace = h5f['kspace'][slice, 0] * np.array(kspace_mask[0])

plt.figure(figsize=(24, 6))

# Plot the real part
plt.subplot(1, 4, 1)
plt.title('Real Part')
plt.imshow(np.log(np.fft.fftshift(np.fft.fft2(masked_kspace[0]))).real, cmap='gray')
plt.colorbar()

# Plot the imaginary part
plt.subplot(1, 4, 2)
plt.title('Imaginary Part')
plt.imshow(np.log(np.fft.fftshift(np.fft.fft2(masked_kspace[0]))).imag, cmap='gray')
plt.colorbar()

# Plot the magnitude
plt.subplot(1, 4, 3)
plt.title('Magnitude')
plt.imshow(np.abs(np.fft.fftshift(np.fft.fft2(masked_kspace[0]))), cmap='gray')
plt.colorbar()

# Plot the phase
plt.subplot(1, 4, 4)
plt.title('Phase')
plt.imshow(np.angle(np.fft.fftshift(np.fft.fft2(masked_kspace[0]))), cmap='gray')
plt.colorbar()

plt.show()
```
