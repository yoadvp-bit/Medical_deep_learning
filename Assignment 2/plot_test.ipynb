```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Data for plotting
# data = pd.read_csv('test_scores_10499706.csv')
# data = pd.read_csv('test_scores_10404727.csv')
data = pd.read_csv('test_scores_10439067.csv')

# Define a color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
markers = ['o', 's', 'D', '^']
```

```python
def show_test_scores(x_data='learning_rate'):
    plt.figure(figsize=(6, 4))

    # Scatter plot with semilogarithmic x-axis
    plt.scatter(data[x_data], data['test_acc'], label='Test Accuracy', alpha=0.5, color=colors[0], marker=markers[0])
    plt.scatter(data[x_data], data['test_f1'], label='Test F1 Score', alpha=0.5, color=colors[1], marker=markers[1])
    plt.scatter(data[x_data], data['test_precision'], label='Test Precision', alpha=0.5, color=colors[2], marker=markers[2])
    plt.scatter(data[x_data], data['test_recall'], label='Test Recall', alpha=0.5, color=colors[3], marker=markers[3])

    if x_data == 'learning_rate':
        plt.xscale('log')

    plt.legend()
    plt.xlabel(x_data.replace('_', ' ').title(), fontsize=14)
    plt.ylabel('Score', fontsize=14)
    plt.title('Test Scores vs ' + x_data.replace('_', ' ').title(), fontsize=16)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```

```python
for column in [col for col in data.columns if not col.startswith('test')]:
    show_test_scores(column)
```
